```python
import matplotlib.pyplot as plot
from sklearn.linear_model import LinearRegression, LogisticRegression 
import seaborn as sns
import numpy as npy
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv("https://raw.githubusercontent.com/pipe2015/Projects_unad/main/data_csv/regresion-lineal.csv", sep=",")
```

```python
data
```

```python
data.plot.scatter(x="metro", y='precio')
plot.show()
```

```python
#Agrego los datos en un array
years_x = data['metro'].values.reshape((-1, 1));
price_y = data['precio'];
#add create model
model = LinearRegression().fit(years_x, price_y);

print('interseccion (b)', model.intercept_)
print('Pendiente (m)', model.coef_)
```

```python
input_list = [[5],[7],[10],[12],[20], [25]]
predicciones = model.predict(input_list)
print(predicciones)
```

```python
data.plot.scatter(x="metro", y="precio", label='Datos originales');
plot.scatter(input_list, predicciones, color='red')
plot.plot(input_list, predicciones, color='black', label='Línea de regresión')
plot.xlabel('metro')
plot.ylabel('precio')
plot.legend()
plot.show()
```
